### Splitting a DataFrame

```python
df=pd.DataFrame({'Name':['Muhammad Adeel','Sohaib Shah','Ahmad Khan','Ali Akbar','Muhammad Umar'],'Location':['Karachi,Pakistan','Lahore,Pakistan','Hyderabad,Pakistan','Karachi,Pakistan','Berlin,Germany']})
df
```

```python
df[['first_name','last_name']]=df.Name.str.split(' ',expand=True)
df
```

```python
df[['city','country']]=df.Location.str.split(',',expand=True)
df
```

```python
df=df[['first_name','last_name','city','country']]
df
```

### Reshaping a DataFrame

```python
fasla=pd.DataFrame([['123',100,200,300],['456',400,500,600],['789',700,800,900]],columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla_2=pd.DataFrame([[1,'123','factory'],[2,'456','warehouse']],columns=['user_id','zip','location_type'])
fasla_2
```

```python
#Converting wide dataset into long dataset
fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
```

```python
fasla_long=fasla.melt(id_vars='zip',var_name='location_type',value_name='distance')
sns.barplot(x='zip',y='distance',hue='location_type',data=fasla_long)
```
